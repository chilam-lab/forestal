```python
import pandas as pd
import numpy as np
```

### Importación de datos

```python
covariables_df = pd.read_csv("sourceData/covariables_sv.csv", dtype={'especievalida':str, 'min':str, 'max':str, 'ET_ID':int})[7407:]
covariables_df = covariables_df[['especievalida', 'ET_ID']]
covariables_df
```

Separando cada variable en diferentes `DataFrames`, desde *Politic07...* hasta *Politic12...*

```python
num_variables: int = int(covariables_df.shape[0] / 2469) # 2469 muncs per variable

variables_dfs = np.empty(num_variables, dtype=pd.DataFrame)
for i in range(num_variables):
    variables_dfs[i] = (covariables_df[['especievalida', 'ET_ID']][i * 2469: (i + 1) * 2469])

for i in range(num_variables):
    print(variables_dfs[i]['especievalida'], ': ', variables_dfs[i].shape)
```

Por último, de cada `DataFrame` se agrupan las celdas con la misma covariable.

```python
def group_by_var(df: pd.DataFrame, variables: str, values: str):
    names = []
    cells = []
    for v in set(df[variables]):
        names.append(v)
        cells.append(df[df[variables]==v][values].tolist())

    return pd.DataFrame({'name':names, 'cells':cells})
```

```python
Politic07_df = group_by_var(variables_dfs[0], 'especievalida', 'ET_ID')
Politic07_df
```

```python
Politic08_df = group_by_var(variables_dfs[1], 'especievalida', 'ET_ID')
Politic08_df
```

```python
Politic09_df = group_by_var(variables_dfs[2], 'especievalida', 'ET_ID')
Politic09_df
```

```python
Politic10_df = group_by_var(variables_dfs[3], 'especievalida', 'ET_ID')
Politic10_df
```

```python
Politic11_df = group_by_var(variables_dfs[4], 'especievalida', 'ET_ID')
Politic11_df
```

```python
Politic12_df = group_by_var(variables_dfs[5], 'especievalida', 'ET_ID')
Politic12_df
```
